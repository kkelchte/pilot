```python
import os
import numpy as np
```

```python
# !!!RUN THIS ONLY ONES!!!!
# parse control and collision info so number start from 0 consecutive to number of images.

root_dir='/esat/opal/kkelchte/docker_home/pilot_data/canyon_rl_turtle'
runs=sorted([root_dir+'/'+d for d in os.listdir(root_dir) if os.path.isdir(root_dir+'/'+d) and '0' in d])
for r in runs:
    print os.path.basename(r)
    for f in ['control_info.txt','control_info_3.txt','collision_info.txt','collision_info_3.txt','collision_info_5.txt','collision_info_7.txt']:
        try:
            fl=open(r+'/'+f,'r')
            lines=fl.readlines()
            fl.close()
        except:
            print("fail to open {}".format(f))
            break
        if 'control' in f: lines=["{0:010d} {1}".format(i,l[11:]) for i,l in enumerate(lines[::2])]
        if 'collision' in f: lines=["{0:010d} {1}".format(i,l[11:]) for i,l in enumerate(lines)]
        try:
            os.rename(r+'/'+f,r+'/'+f+'_old')
            fl=open(r+'/'+f,'w')
            for l in lines: fl.write(l)
            fl.close()
        except:
            print("fail to write {}".format(f))
            break
```
